```python
import pandas as pd
%matplotlib inline
```

```python
filepath = '../data/all-stop-and-search-2015.csv'
```

```python
df = pd.read_csv(filepath, usecols=['Type', 'Date', 'Age range', 'Officer-defined ethnicity', 'Object of search', 'Outcome', 'Removal of more than just outer clothing'])
```

```python
total_num_stops = df.groupby(['Officer-defined ethnicity', 'Type']).count()['Date'].sum()
```

```python
t = df.groupby(['Officer-defined ethnicity', 'Type']).count()['Date'] / total_num_stops * 100
t.plot.bar()
```

```python
t = df.groupby(['Age range', 'Type']).count()['Date'] / total_num_stops * 100
t.plot.bar()
```
