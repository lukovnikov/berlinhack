```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv("../data/local-areas-of-ethnic-group-concentration.csv")
```

```python
df = df.drop("Borough", 1)
```

```python
for ethnicity in "White British;White Irish;White Other;Mixed;Asian;Black;Other".split(";"):
    df[ethnicity] = df[ethnicity].map(lambda x: float(x[:-1])/100.)
```

```python
df
```

```python
# add up whites
df["White"] = df["White British"] + df["White Irish"] + df["White Other"]
```

```python
df = df.drop(["White British", "White Irish", "White Other"], 1)
```
